```python
import cv2
from matplotlib import pyplot as plt

image = cv2.imread("C:\\Users\\Robin\Desktop\\ocrinaction-master\\ocrinaction-master\\chapter-2\\img\\2-1.png")
#

print(image.shape)
# NumPy  channelRGB  channel  3 1

gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
# image 

plt.figure(figsize=(100,50))
plt.subplot(311), plt.imshow(gray, "gray")
plt.title("input image"), plt.xticks([]), plt.yticks([])
#

ret1, th1 = cv2.threshold(gray, 0, 255, cv2.THRESH_OTSU)
#(ostu)

plt.subplot(312), plt.hist(gray.ravel(), 256)
#

plt.axvline(x=ret1, color='green', label='otsu')
plt.legend(loc='upper right')
plt.title("Histogram"), plt.xticks([]), plt.yticks([])
#ostu

plt.figure(figsize=(20,10))
plt.subplot(313), plt.imshow(th1, "gray")
plt.title("output image"), plt.xticks([]), plt.yticks([])
plt.show()
#
```
